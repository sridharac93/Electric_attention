```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('Electric_fault.csv')
```

```python
df.head()
```

```python
x = df[['Ia','Ia','Ic','Va','Vb','Vc']]
y = df[['G','C','B','A']]
```

```python
l = []
for i in range(0,len(y)):
    l.append([])
    l[i].append(y.iloc[i,:][0])
    l[i].append(y.iloc[i,:][1])
    l[i].append(y.iloc[i,:][2])
    l[i].append(y.iloc[i,:][3])
```

```python
len(l)
```

```python
uniq = [[1,0,0,1]]
for i in range(1,len(l)):
    if l[i] not in uniq:
        uniq.append(l[i])
```

```python
uniq
```

```python
s = {'LG':[1, 0, 0, 1],'LLG':[1, 0, 1, 1],'LL':[0, 1, 1, 0],'LLL':[0, 1, 1, 1],'LLLG':[1, 1, 1, 1],'No_F':[0, 0, 0, 0]}
```

```python
for i in range(len(l)):
    for k,v in s.items(): 
        if l[i] == v :
            l[i] = k
            
```

```python
l[-1]
```

```python
df['faults'] = l
```

```python
df['faults'].unique()
```

```python
x = df[['Ia','Ia','Ic','Va','Vb','Vc']]
y = df['faults']
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
y = tf.keras.utils.to_categorical(y)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2)
```

```python
model = tf.keras.Sequential()
```

```python
model.add(tf.keras.layers.Dense(128,activation = 'relu', input_shape = (6,)))
model.add(tf.keras.layers.Dense(256,activation = 'relu'))
model.add(tf.keras.layers.Dense(128,activation = 'relu'))
model.add(tf.keras.layers.Dense(6,activation = 'softmax'))
```

```python
model.compile(loss = 'categorical_crossentropy',optimizer = 'adam',metrics = ['acc'])
```

```python
model.fit(x_train,y_train,epochs = 50)
```
